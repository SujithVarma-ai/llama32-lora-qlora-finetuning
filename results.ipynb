## Let's plot a bar graph of all the models error rate 

```python
import plotly.graph_objects as go

results = [
    ("Constant", "blue", 106.08),
    ("Linear Regression", "yellow", 92.31),
    ("NLP + LR", "brown", 79.67),
    ("Random Forest", "skyblue", 71.57),
    ("XGBoost", "gray", 76.63),
    ("Human (Ed)", "black", 87.62),
    ("Neural Network", "orange", 74.54),
    ("GPT 4.1 Nano", "slateblue", 58.23),
    ("Gemini 2.5 Flash Lite", "purple", 50.54),
    ("GPT 5.1", "green", 41.74),
    ("Base Llama 3.2 4 bit", "orange", 106.23),
    ("Fine-tuned smaller dataset","blue", 63.82),
    ("Fine-tuned larger dataset", "green", 39.88)
]

labels, colors, values = zip(*results)

fig = go.Figure(go.Bar(x=labels, y=values, marker_color=colors))

fig.update_layout(
    title="Week 6 model improvements - see the videos to reveal which triumphs!",
    yaxis=dict(range=[0, max(values)], title="Error"),
    xaxis=dict(tickangle=-45),
    width=1000,
    height=800
)

fig.show()
```
